```
#Creation de masks
look_for = ["attr1_1","sinc1_1","intel1_1","fun1_1","amb1_1","shar1_1","gender","iid"]
opp_look_for = ["attr2_1","sinc2_1","fun2_1","amb2_1"]
own_score = ["attr3_1","sinc3_1","fun3_1","amb3_1"]
think_own_score = ["attr5_1","sinc5_1","fun5_1","amb5_1"]

looking_dataset = dataset[look_for]
looking_dataset
```

```
look_values = looking_dataset.iloc[0:8378,0:6]
look_values
```

```
homme=looking_dataset.gender==0
femme=looking_dataset.gender==1

labels=["Attractif","Sinc√®re","Intello","Drole","Ambitieux","Meme delires"]
x = np.arange(len(labels))
width = 0.35

femme_means = look_values[femme].mean()
homme_means = look_values[homme].mean()
```

```
fig, ax = plt.subplots()
men = ax.bar(x - width/2, homme_means, width=0.35, label='Men')
women = ax.bar(x + width/2, femme_means, width=0.35, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel("% Interet")
ax.set_title("Ce que recherche les Femmes VS Hommes")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```
